### 1️⃣ **대한민국의 노인 인구 증가 추이 분석**
1990년부터 현재까지의 65세 이상 노인 인구 구성비 변화 추이를 시각화하여 분석하고, 2070년까지의 노인 인구 추정 구성비까지 나타내 고령화의 정도와 속도를 파악한다.

- 데이터 수집
: 통계청,「장래인구추계」, 2070, 2023.06.18, 주요 인구지표(성비,인구성장률,인구구조,부양비 등) / 전국
(https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1BPA002&conn_path=I2)

```python
import pandas as pd

df = pd.read_csv('/content/주요_인구지표_성비_인구성장률_인구구조_부양비_등.csv', encoding='cp949')
df.info()
```

```python
df
```

```python
ratio_65_older = df[df['인구구조,부양비별'] == '- 구성비(%): 65세 이상']
ratio_65_older.drop('가정별', axis=1, inplace=True) # '가정별' 열 삭제
ratio_65_older
```

```python
# '인구구조,부양비별' 열을 인덱스로 설정하고, 열 이름을 '기준'으로 변경
ratio_65_older.set_index('인구구조,부양비별', inplace=True)
ratio_65_older.index.name = '기준'
ratio_65_older
```

```python
ratio_65_older.index
```

```python
ratio_65_older = ratio_65_older.rename(index={'- 구성비(%): 65세 이상': '65세이상'})
ratio_65_older
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 선 그래프 그리기
plt.plot(ratio_65_older.columns, ratio_65_older.loc['65세이상'])

# 그래프 제목 및 축 레이블 설정
plt.title('65세 이상 구성비(%) 변화')
plt.xlabel('연도')
plt.ylabel('구성비(%)')

# x축 눈금 라벨 회전하기
plt.xticks(rotation=90)

# 수직선 추가 - 2023년 표시
plt.axvline(x='2023', color='red', linestyle='--')


# 그래프 표시
plt.show()
```

💡 **고령화 추이 및 예측** 💡
<br>대한민국의 노인 인구 증가 추이 분석에 대한 시각화를 통해, 1990년부터 2023년까지 65세 이상의 고령인구 구성비가 20%까지 증가하였음을 알 수 있다. 또한, 점점 증가 속도가 빨라져 2050년 이후에는 65세 이상의 고령 인구 구성비가 40%가 넘을 것으로 전망된다.

### 2️⃣ **대한민국의 노인 인구 경제 활동 상태**
2005년부터 2022까지의 65세 이상 노인 인구의 경제 활동 상태에 관해 시각화하고, 어떤 문제점이 있는지 분석한다.

- 데이터 수집
: 통계청,「경제활동인구조사」, 2022.05, 2023.06.18, 연령별 경제활동상태
(https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1DE8031S&conn_path=I2)

```python
df = pd.read_csv('/content/연령별_경제활동상태.csv', encoding='cp949')
df.info()
```

```python
df
```

```python
# '연령별' 열 삭제
df.drop('연령별', axis=1, inplace=True)

# '인구구조,부양비별' 열을 인덱스로 설정하고, 열 이름을 '기준'으로 변경
df.set_index('항목', inplace=True)
df
```

```python
df = df.transpose()
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# 데이터프레임 설정
df_plot = df

# 선 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

# 왼쪽 축 - '고령층인구', '경제활동인구', '취업자', '실업자', '비경제활동인구'
df_plot[['고령층인구', '경제활동인구', '취업자', '실업자', '비경제활동인구']].plot(kind='line', ax=ax1)
ax1.set_ylabel('명', rotation=360)

# 오른쪽 축 - '고용률 (%)', '실업률 (%)'
df_plot[['실업률 (%)']].plot(kind='line', ax=ax2, linestyle='--', color='b')
df_plot[['고용률 (%)']].plot(kind='line', ax=ax2, linestyle='--', color='r')
ax2.set_ylabel('%', rotation=360)

# 그래프 제목과 범례 설정
plt.title('연도별 데이터 변화')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 축 레이블 설정
ax1.set_xlabel('연도')

# 그래프 출력
plt.show()
```

💡 **경제 활동 상태 현황** 💡
<br>노인 인구가 점차 증가함에 따라 대체로 경제활동인구, 취업자, 비경제활동인구는 증가하는 경향을 보이고 있다. 반면 실업자 수는 약간의 변동이 있었지만 전체적으로 감소하는 경향을 보인다. 고용률은 연도별로 변동이 있으나 전반적으로 증가하는 추세를 보이고, 실업률은 연도에 따라 변동이 있지만 전체적으로 감소하는 경향을 보인다.

💡 **노인 인구 증가와 경제 활동 상태의 관계** 💡
<br>노인 인구의 증가율과 경제 활동 상태를 비교해보면, 노인 인구의 증가와 고령층인구, 경제활동인구, 취업자 수가 증가하는 경향을 보이며, 이는 노인들의 경제 활동 참여가 증가하고 있다는 것을 시사한다. 즉, 노인들이 적극적으로 경제 활동에 참여하고 있으며, 일자리에 대한 수요와 맞물려 실업률이 감소하고 있는 것을 의미한다.

💡 **문제점** 💡
<br>노인 인구의 증가에 비해 실업자 수는 감소하는 경향을 보이지만, 일자리의 질과 안정성에 대한 문제가 있을 수 있으며, 일부 노인들은 저임금이나 불규칙한 일자리를 갖는 경우가 있을 수 있다. 이를 개선하기 위해 직업 훈련 및 적절한 일자리 창출이 필요하다.  또한, 절대적인 비경제활동인구는 늘었기 때문에 이러한 노인 인구에 대한 사회적 정책이나 조치가 필요하다고 생각된다.

### 3️⃣ **대한민국의 노인 질병 유병률**
2020년 기준 노인에게 발병한 질병에 대해 조사하고, 성별에 따라 어떤 차이가 있는지 분석한다.

- 데이터 수집
: 보건복지부,「노인실태조사」, 2020, 2023.06.18, 노인의 성별 만성질병 종류별 유병률(의사진단 기준) 및 현 치료율 (https://kosis.kr/statHtml/statHtml.do?orgId=117&tblId=DT_117071_018&conn_path=I2)

```python
df = pd.read_csv('/content/노인의_성별_만성질병_종류별_유병률.csv', encoding='cp949')
df.info()
```

```python
df
```

```python
df = df.rename(columns=df.iloc[0])
df = df.drop(df.index[0])
df = df.drop(df.index[0])
df
```

```python
# 유병률 데이터 자료형을 object에서 float으로 변경
df['전체'] = df['전체'].astype(float)
df['남자'] = df['남자'].astype(float)
df['여자'] = df['여자'].astype(float)
df.info()
```

```python
import matplotlib.pyplot as plt

# 인덱스 지정
df = df.set_index('만성질환별(1)')

# 그래프 그리기
df.plot(kind='bar', figsize=(12, 6))
plt.xlabel('만성질환')
plt.ylabel('발생률 (%)')
plt.title('2020년 만성질환 발생률')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.show()
```

💡**발생률이 가장 높은 질환**💡
<br> 고혈압이 전체 56.8%로 가장 높은 발생률을 가지고, 당뇨병은 24.2%로 두 번째로 높은 발생률을 가지고 있다. 성별에 따른 차이는 존재하지만 큰 차이는 아니며, 둘 다 중요한 만성질환으로 간주된다. 따라서 고혈압과 당뇨병에 대한 예방과 조기 치료에 주의를 기울이는 것이 중요하다.

💡**여성 질환**💡
<br> 남성보다 여성에게 더 많이 발생되는 것으로 보이는 질환에는 골관절염, 류머티즘관절염, 골다공증, 요통, 좌골신경통, 요실금, 빈혈 등이 있다. 여성은 특히 이러한 질병에 주의를 기울여야 할 것으로 보인다.

💡**남성 질환**💡
<br> 여성보다 남성에게 더 많이 발생되는 것으로 보이는 질환에는 전립성비대증이 있고, 그 외의 질환에서는 남성과 여성 발생율에 큰 차이가 없는 것으로 보인다.
